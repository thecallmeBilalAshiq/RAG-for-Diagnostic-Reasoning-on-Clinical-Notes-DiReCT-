It's a method used in NLP (Natural Language Processing) to improve how AI can generate answers or responses. Here's how it works in simple terms:

- 1️⃣ Retrieval: The AI looks for relevant information from a large set of data, like documents or articles.

- 2️⃣ Augmented: It uses this retrieved information to help improve the answer.

- 3️⃣ Generation: The AI then creates a response based on the retrieved info.

So, RAG helps the AI generate more accurate answers by first looking up related data and then using that information to create a better response. It's like asking a smart assistant to read a few pages before answering your question.

# 2️⃣ Data Loading and Filtering

*   This processes clinical documents from the MIMIC-IV-Ext dataset by extracting key information, organizing it into a **structured DataFrame(Panda)**, and filtering out incomplete records for use in a Retrieval-Augmented Generation (RAG) system.





```python
import os
import json
import pandas as pd

class MIMICDataLoader:
    def __init__(self):
        pass

    def load_and_preprocess(self, data_path):
        all_notes = []

        for root, dirs, files in os.walk(data_path):
            for file in files:
                if file.endswith('.json'):
                    file_path = os.path.join(root, file)
                    try:
                        with open(file_path, 'r') as f:
                            data = json.load(f)

                            if len(all_notes) < 6:
                                print(f"\nInspecting {file}")
                                print(f"Available Keys: {list(data.keys())}")
                                print(f"Diagnosis: {data.get('Diagnosis', 'N/A')}")
                                print(f"Treatment: {data.get('Treatment', 'N/A')}\n")

                            keys = list(data.keys())
                            if not keys:
                                continue

                            # Extract diagnosis information from the first key
                            main_key = keys[0]
                            diagnosis_raw = main_key.split('$')[0].replace('_', ' ')
                            diagnosis = diagnosis_raw if diagnosis_raw else 'Unknown'

                            # Gather relevant information from various inputs
                            chief_complaint = data.get('input1', 'No complaint recorded')
                            history = data.get('input2', 'No history provided')
                            treatment = data.get('input3', 'No treatment data')

                            document_text_parts = [data.get(f'input{i}', '') for i in range(1, 7)]
                            document_text = f"Diagnosis: {diagnosis} | Complaint: {chief_complaint} | History: {history} | Treatment: {treatment} | Notes: {' '.join(document_text_parts)}"

                            # Add the extracted data to the list for future processing
                            all_notes.append({
                                'document_id': file,
                                'document_text': document_text,
                                'chief_complaint': chief_complaint,
                                'history': history,
                                'diagnosis': diagnosis,
                                'treatment': treatment
                            })
                    except json.JSONDecodeError:
                        print(f"Error decoding JSON from file: {file_path}")
                    except Exception as e:
                        print(f"Error processing file: {file_path}, Details: {e}")

        # Return the notes as a DataFrame for easy manipulation
        return pd.DataFrame(all_notes)

# Instantiate the data loader object
data_loader = MIMICDataLoader()

# Define the path to the unzipped MIMIC-IV directory
data_path = '/content/mimic-iv-ext-direct-1.0.0/' # Corrected path after unzipping

# Load and preprocess the data using the defined method
original_notes_df = data_loader.load_and_preprocess(data_path)

# Filter out documents where diagnosis or treatment fields are empty
# The filtering logic ensures that only relevant documents are retained
notes_df = original_notes_df[(original_notes_df['diagnosis'].str.strip() != '') | (original_notes_df['treatment'].str.strip() != '')]

# Output information about the document processing
print(f"Total number of documents processed: {len(original_notes_df)}")
print(f"Number of documents after filtering: {len(notes_df)}")
print("\nData processing and filtering complete. Here's a preview of the filtered data:")
print(notes_df.head())
```

# 3️⃣ Designing the Retrieval Component  ✅✅
- Retriever Selection: Choose or implement a retriever (e.g., BM25 or dense
retrieval using embeddings) to index and search the dataset eTectively.  

- Indexing Strategy: Develop an indexing strategy that handles the volume of
clinical records.  

- Query Formulation: Create a method for converting natural language clinical
queries into eTective search queries for your retrieval engine.
---

- Setting up a **similarity search systemusing FAISS** (Facebook AI Similarity Search) to retrieve the most relevant clinical documents based on a given query. The core idea is to encode the documents into dense vector embeddings (numerical representations of text) using a SentenceTransformer model, and then use FAISS to search and retrieve the most similar documents quickly.
----
1. FAISS is used for efficient similarity search, enabling fast retrieval of similar documents.

2. SentenceTransformer is used to encode documents and queries into vector embeddings.

3. The FAISS index stores these embeddings and allows for fast similarity-based searches.


```python
# Install FAISS for efficient similarity search
!pip install faiss-cpu

# Import necessary libraries
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

class ClinicalRetriever:
    def __init__(self):
        # Load the pre-trained SentenceTransformer model for encoding text
        self.model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
        # Initialize index and metadata placeholders
        self.index = None
        self.documents = None
        self.metadata = None

    def build_index(self, documents, metadata):
        """
        Build the FAISS index with the provided documents and metadata.
        This method encodes the documents using a sentence transformer model,
        normalizes the embeddings, and adds them to the FAISS index.
        """
        self.documents = documents
        self.metadata = metadata

        # Encode the documents into dense vector embeddings
        embeddings = self.model.encode(documents, show_progress_bar=True, batch_size=32)
        embeddings = np.array(embeddings).astype('float32')  # Ensure the embeddings are in float32 format

        # Initialize the FAISS index for inner product search
        self.index = faiss.IndexFlatIP(embeddings.shape[1])

        # Normalize the embeddings to unit vectors for cosine similarity
        faiss.normalize_L2(embeddings)

        # Add embeddings to the FAISS index
        self.index.add(embeddings)

    def retrieve(self, query, top_k=5):
        """
        Retrieve the top-k most similar documents for a given query.
        This method encodes the query, normalizes the query embedding,
        and searches the FAISS index for the most similar documents.
        """
        # Encode the query into a dense vector embedding
        query_emb = self.model.encode([query])
        query_emb = np.array(query_emb).astype('float32')  # Ensure the query embedding is in float32 format

        # Normalize the query embedding
        faiss.normalize_L2(query_emb)

        # Perform the similarity search to find the top_k most relevant documents
        scores, indices = self.index.search(query_emb, top_k)

        # Return the top-k documents with their metadata and similarity score
        return [(self.documents[idx], self.metadata[idx], float(scores[0][i]))
                for i, idx in enumerate(indices[0]) if idx < len(self.documents)]

# Instantiate the ClinicalRetriever object
retriever = ClinicalRetriever()

# Build the index using the preprocessed documents from the previous steps
retriever.build_index(notes_df['document_text'].tolist(), notes_df.to_dict('records'))

print("Index built successfully. You can now perform retrieval queries.")
```

```python
from datetime import datetime

class ClinicalGenerator:
    def generate_response(self, query, retrieved_docs):
        response = f"----> Query: '{query}'\n\n"
        for i, (doc, meta, score) in enumerate(retrieved_docs, 1):
            response += f"{i}. Relevance: {score:.2f}\n"
            if 'diagnosis' in meta:
                response += f"   Diagnosis: {meta['diagnosis']}\n"
            if 'treatment' in meta:
                response += f"   Treatment: {meta['treatment'][:100]}...\n\n"
        return response

generator = ClinicalGenerator()

class ClinicalRAGSystem:
    def __init__(self, retriever, generator, df):
        self.retriever = retriever
        self.generator = generator
        self.df = df

    def query(self, query, top_k=3):
        docs = self.retriever.retrieve(query, top_k)
        response = self.generator.generate_response(query, docs)
        return {
            'query': query,
            'response': response,
            'retrieved_documents': [{'text': d, 'metadata': m, 'score': s} for d, m, s in docs],
            'timestamp': datetime.now().isoformat()
        }

rag_system = ClinicalRAGSystem(retriever, generator, notes_df)

queries = [
    "liver failure treatment",
    "hypertension diagnosis",
    "stroke symptoms",
    "seizure management",
    "sepsis treatment",
    "heart attack signs",
    "kidney stone symptoms",
    "COPD exacerbation",
    "asthma management"
]

for q in queries:
    result = rag_system.query(q, 3)
    print(result['response'])
    print("_"*80 + "\n")
```

```python
!pip install -q transformers torch

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

class HuggingFaceGenerator:
    def __init__(self):
        self.tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
        self.model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large")
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        self.model.to(self.device)

    def generate_response(self, query, retrieved_docs):
        context = "\n".join([f"Doc{i}: {doc[:200]}" for i, (doc,_,_) in enumerate(retrieved_docs,1)])
        prompt = f"Answer medical question.\n\nRecords: {context}\n\nQuestion: {query}\n\nAnswer:"
        inputs = self.tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True).to(self.device)
        outputs = self.model.generate(**inputs, max_length=300, num_beams=4)
        return self.tokenizer.decode(outputs[0], skip_special_tokens=True)

    def generate_pure_text(self, prompt_text, max_length=150, num_beams=4):
        inputs = self.tokenizer(prompt_text, return_tensors="pt", max_length=512, truncation=True).to(self.device)
        outputs = self.model.generate(**inputs, max_length=max_length, num_beams=num_beams)
        return self.tokenizer.decode(outputs[0], skip_special_tokens=True)

generator = HuggingFaceGenerator()


from datetime import datetime

class EnhancedRAGSystem:
    def __init__(self, retriever, generator, df):
        self.retriever = retriever
        self.generator = generator
        self.df = df

    def query(self, query, top_k=3):
        docs = self.retriever.retrieve(query, top_k)
        response = self.generator.generate_response(query, docs)
        return {
            'query': query,
            'response': response,
            'retrieved_documents': [{'text': d, 'metadata': m, 'score': s} for d,m,s in docs],
            'timestamp': datetime.now().isoformat()
        }

enhanced_rag = EnhancedRAGSystem(retriever, generator, notes_df)

queries = ["diabetes treatment", "stroke symptoms", "asthma management", "sepsis treatment"]

for q in queries:
    result = enhanced_rag.query(q, 3)
    print(f"Query: {q}\n\n{result['response']}")
    print("_______________" + "\n")
```

```python
import gradio as gr
from datetime import datetime

class EnhancedRAGSystem:
    def __init__(self, retriever, generator, df):
        self.retriever = retriever
        self.generator = generator
        self.df = df

    def query(self, query, top_k=3):
        docs = self.retriever.retrieve(query, top_k)
        response = self.generator.generate_response(query, docs)
        return {
            'query': query,
            'response': response,
            'retrieved_documents': [
                {'text': d, 'metadata': m, 'score': s} for d, m, s in docs
            ],
            'timestamp': datetime.now().isoformat()
        }

enhanced_rag = EnhancedRAGSystem(retriever, generator, notes_df)


def normalize_query(user_input):
    user_input = user_input.lower()

    if "diabetes" in user_input:
        return "diabetes treatment"
    if "stroke" in user_input:
        return "stroke symptoms"
    if "asthma" in user_input:
    	return "asthma management"
    if "sepsis" in user_input:
        return "sepsis treatment"


    return user_input


def rag_pipeline(user_input):

    normalized = normalize_query(user_input)
    result = enhanced_rag.query(normalized, top_k=3)

    disease_name = "N/A"
    if result["retrieved_documents"] and 'diagnosis' in result["retrieved_documents"][0]['metadata']:
        disease_name = result["retrieved_documents"][0]['metadata']['diagnosis']

    final_answer = f"""
### 🔍 Normalized Query:
**{normalized}**

### 💡 Answer:
**Disease:** {disease_name}
{result['response']}

---

### 📄 Most Relevant Documents:
"""
    if result["retrieved_documents"]:
        for i, doc in enumerate(result["retrieved_documents"]):
            final_answer += f"""
**{i+1}. Score:** {doc['score']:.4f}
**Text:** {doc['text'][:300]}...

"""

    return final_answer

ui = gr.Interface(
    fn=rag_pipeline,
    inputs=gr.Textbox(lines=3, label="Ask a question"),
    outputs=gr.Markdown(label="RAG Answer"),
    title="Medical RAG System by Bilal Ashiq",
    description="Type any medical question. The system will retrieve notes, and generate an answer."
)

ui.launch()
```

```python
from sentence_transformers import util

def evaluate_retrieval_semantic(query, retrieved_docs, full_df, embed_model, threshold=0.40):


    q_emb = embed_model.encode(query, convert_to_tensor=True)


    all_embs = embed_model.encode(full_df["clean_text"].tolist(), convert_to_tensor=True)
    sims = util.cos_sim(q_emb, all_embs)[0]

    relevant_indices = (sims > threshold).nonzero().flatten().tolist()
    total_relevant = len(relevant_indices)

    retrieved_indices = [d[1]["row_id"] for d in retrieved_docs]
    relevant_retrieved = len(set(retrieved_indices) & set(relevant_indices))

    precision = relevant_retrieved / len(retrieved_docs)
    recall = relevant_retrieved / total_relevant if total_relevant > 0 else 0

    return precision, recall
```

```python
def evaluate_answer(query, answer, generator):

    prompt_for_evaluation = f"""
Rate the clinical response on a scale of 1—5.

Query: {query}
Response: {answer}

Assessment criteria:
Relevance (1-5)
Coherence (1-5)
Accuracy (1-5)

Provide your response exactly like this:
Relevance: X
Coherence: Y
Accuracy: Z
"""

    result = generator.generate_pure_text(prompt_for_evaluation, max_length=100)

    print(f"\n Model's QEval '{query[:50]}' ---")
    print(result)

    import re
    relevance_match = re.search(r"Relevance:\s*(\d)", result, re.IGNORECASE)
    coherence_match = re.search(r"Coherence:\s*(\d)", result, re.IGNORECASE)
    accuracy_match = re.search(r"Accuracy:\s*(\d)", result, re.IGNORECASE)

    if relevance_match and coherence_match and accuracy_match:
        return int(relevance_match.group(1)), int(coherence_match.group(1)), int(accuracy_match.group(1))

    return (3, 3, 3)
```

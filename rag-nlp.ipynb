```python
# --- CELL 2: DATA INGESTION (Run AFTER Restart) ---
import os
import zipfile

# --- DEBUG CHECK ---
try:
    from transformers import AutoModel
    from sentence_transformers import SentenceTransformer
    print("вњ… Libraries are working perfectly now!")
except ImportError as e:
    print(f"вќЊ STILL BROKEN: {e}")
    # Stop here if broken
    raise e

from langchain_community.document_loaders import DirectoryLoader, TextLoader
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 1. SETUP PATHS
input_root = '/kaggle/input'
working_dir = '/kaggle/working/dataset'
persist_directory = "/kaggle/working/chroma_db"

# 2. FIND DATA
data_source_path = None
# Check for Zip
for root, dirs, files in os.walk(input_root):
    for file in files:
        if file.endswith("dataset.zip"):
            zip_path = os.path.join(root, file)
            if not os.path.exists(working_dir):
                with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                    zip_ref.extractall(working_dir)
            data_source_path = working_dir
            break
    if data_source_path: break

# Check for Folder
if not data_source_path:
    for root, dirs, files in os.walk(input_root):
        if len(files) > 0 and not root.startswith("/kaggle/input/kernel-"):
            data_source_path = root
            break

# 3. RUN PIPELINE
if data_source_path:
    print(f"вЏі Processing data from: {data_source_path}")
    loader = DirectoryLoader(data_source_path, glob="**/*.txt", loader_cls=TextLoader)
    documents = loader.load()
    
    if documents:
        print(f"рџ“„ Loaded {len(documents)} documents.")
        text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
        texts = text_splitter.split_documents(documents)
        
        print("рџ’ѕ Generating Embeddings...")
        # Using the specific compatible model
        embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
        
        vectordb = Chroma.from_documents(
            documents=texts,
            embedding=embedding_model,
            persist_directory=persist_directory
        )
        print(f"вњ… SUCCESS! Database saved to: {persist_directory}")
    else:
        print("вљ пёЏ No documents found.")
else:
    print("вќЊ dataset.zip not found.")
```

```python
# --- STEP 3: RUN VIVA TEST ---
import sys
import torch
import time

# Verify imports before crashing
try:
    from langchain.chains import RetrievalQA
    from langchain_community.embeddings import HuggingFaceEmbeddings
    from langchain_community.vectorstores import Chroma
    from langchain_community.llms import HuggingFacePipeline
    print("вњ… SUCCESS: All libraries are locked and loaded!")
except ImportError as e:
    print(f"вќЊ STILL BROKEN: {e}")
    print("рџ‘‰ You did not run Step 1 or did not Restart properly.")
    sys.exit()

from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig
from langchain.prompts import PromptTemplate

print("рџ§Є STARTING FINAL SYSTEM TEST...")

# 1. Load Database
print("рџ“‚ Loading Database...")
try:
    embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    vectordb = Chroma(persist_directory="/kaggle/working/chroma_db", embedding_function=embedding_model)
    print(f"вњ… Database Loaded. Docs found: {vectordb._collection.count()}")
except Exception as e:
    print(f"вќЊ Database Error: {e}")
    sys.exit()

# 2. Load AI
print("рџ§  Loading Mistral-7B...")
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
)

model_id = "mistralai/Mistral-7B-Instruct-v0.2"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto"
)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=256,
    temperature=0.1,
    repetition_penalty=1.1
)
llm = HuggingFacePipeline(pipeline=pipe)

# 3. Ask Question
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vectordb.as_retriever(search_kwargs={"k": 2})
)

print("\n" + "="*40)
query = "What is the diagnosis for the patient?"
print(f"вќ“ Q: {query}")
res = qa_chain.invoke({"query": query})
print(f"рџ¤– A: {res['result'].strip()}")
print("="*40)
print("вњ… SYSTEM READY FOR VIVA.")
```
